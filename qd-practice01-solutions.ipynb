### Задача 2

Напишите программу, которая сохраняет рост пользователя в сантиметрах (переменная `height`), его вес в килограммах (переменная `weight`) и выводит на экран сообщение вида: `Индекс массы тела: [значение].`, где вместо `[значение]` подставляется посчитанное значение индекса массы тела, округлённое до второго знака после запятой.

Индекс массы тела считается так:

$$
BMI=\frac{m}{h^2},
$$

где $m$ – масса тела в килограммах, $h$ – рост в метрах.

**Пример работы программы:**

На входе:
    
    рост: 168
    вес: 55
    
Результат на экране:
    
    Индекс массы тела: 19.49

```python
height = 168
weight = 55
r = weight / (height / 100) ** 2
print("Индекс массы тела:", round(r, 2))
```

### Задача 3

Используя код из предыдущей задачи, напишите новую программу, которая выводит на экран значение `True`, если индекс массы тела в норме (значения от 18.5 до 25), и значение `False`, если индекс массы тела выходит за рамки нормы в меньшую или большую сторону.

### Задача 4

Напишите программу, которая сохраняет год рождения пользователя (переменная `year`, целое число), информацию о том, был ли у того день рождения в этом году или нет (переменная `bd`, значения `True` или `False`), и выводит на экран число полных лет этого пользователя.

*Подсказка:* Python умеет автоматически превращать `True` в 1, а `False` в 0, если эти значения встречаются в операциях с числами.

**Пример работы программы:**

На входе:
    
    1994
    True
    
На экране:

    30
    
На входе:
    
    1957
    False
    
На экране:

    66

```python
# хитрость: считаем число лет на прошлый год,
# а потом добавляем bd (False = 0, True = 1),
# кому надо год добавляется, кому не надо - нет

year = 1994
bd = True
age = 2023 - year + bd
print(age)
```

```python
# если изменяем year и bd, все ок

year = 1957
bd = False
age = 2023 - year + bd
print(age)
```

### Задача 5

Используя код из предыдущей задачи, напишите программу, которая:

а) выводит на экран значение `True`, если пользователь является совершеннолетним, `False` – иначе;

б) сохраняет в переменную `pol` пол респондента (значения `муж` и `жен`) и выводит на экран значение `True`, если пользователь является несовершеннолетним или пенсионером (для женщин с 60 лет, для мужчин – с 65 лет).

```python
# a
print(age >= 18)
```

```python
# b: в age сейчас значение 66

pol = "муж"

# пол и возраст для пенсии объединяем через AND
# итого через ИЛИ три условия: 
# менее 18 лет | мужчины от 65 и старше | женщины от 60 и старше

print((age < 18) | ((pol == "муж") & (age >= 65)) | (pol == "жен") & (age >= 60))
```

### Задача 6

В среднем за неделю Питон получает пять сообщений от Анаконды ($\lambda$=5). В переменную `k` сохраняется число сообщений, которые Анаконда может прислать Питону. Напишите программу, которая выводит на экран вероятность, с которой Питон получит $k$
сообщений от Анаконды за неделю, с точностью до трёх знаков после запятой. Сообщение, выводимое на экран, должно быть такого вида:

    Число сообщений от Анаконды за неделю равно [k], вероятность равна [значение]

Вероятность того, что Питон получит ровно $k$ сообщений, определяется следующим образом (распределение Пуассона, кто знаком с основами теории вероятностей):

$$
P(X=k)=e^{−\lambda}⋅\frac{\lambda^k}{k!}.
$$

Число `e` и функцию `factorial()` для вычисления факториала ($k!$) можно вызвать из модуля `math`.

**Пример работы программы:**

На входе:

    2

Вывод:

    Число сообщений от Анаконды за неделю равно 2, вероятность равна 0.084

```python
import math
```

```python
# вычисляем вероятность

Lambda = 5
k = 2
p = round(math.e ** (-Lambda) * Lambda ** k / math.factorial(k), 3)
print(k, p)
```

```python
# красивый вариант подстановки k и p в строку

print(f"Число сообщений от Анаконды за неделю равно {k}, вероятность равна {p}")
```

**Дополнение для желающих.** Можем написать цикл, который будет перебирать значения от 0 до 10 и вычислять для каждого $k$ от 0 до 10 вероятность:

```python
# правый конец в range() не включается

for k in range(0, 11):
    p = round(math.e ** (-Lambda) * Lambda ** k / math.factorial(k), 3)
    print(f"Число сообщений от Анаконды за неделю равно {k}, вероятность равна {p}")
```
