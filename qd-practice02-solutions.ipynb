*Подсказка:* функции `min()`, `max()`, `mean()` и `median()` из `np`; функция `quantile()` из `np`, на втором месте нужно указать уровень квантиля в виде дробного числа.

```python
min_ = np.min(scores)
max_ = np.max(scores)
mean_ = np.mean(scores)
median_ = np.median(scores)

q1 = np.quantile(scores, 0.25)
q3 = np.quantile(scores, 0.75)

print("Минимум и максимум:", min_, max_)
print("Среднее:", mean_)
print("Медиана:", median_)
print("Нижний и верхний квартили:", q1, q3)
```

>**Комментарии.** В выборке есть организация (или несколько организаций) с максимальным значением индекса 100 из 100, то есть «образцовая» с точки зрения качества работы некоммерческая организация. ### Задача 1.3

Используя посчитанные в предыдущей задаче описательные статистики, вычислите границы типичных значений и сохраните их в переменные `lower` и `upper`. Выберите значения из `scores`, которые являются нетипичными, и сохраните их в массив `out`. Посчитайте их количество.

```python
ir = q3 - q1
lower = q1 - 1.5 * ir
upper = q3 + 1.5 * ir

# 5 значений
out = scores[(scores < lower) | (scores > upper)]
print(len(out), out)
```

### Задача 1.4

Выберите из `scores` типичные значения и сохраните их в массив `scores_typ`. Вычислите среднее и медиану выборки, сохранённой в `scores_typ`. Сравните полученные результаты со средним и медианой из задачи 1.2. Прокомментируйте сходства/различия.

```python
# вариант 1
scores_typ = scores[((scores >= lower) & (scores <= upper))]
```

```python
# вариант 2
# берем условие для нетипичных из задачи 1.3
# добавляем вокруг него круглые скобки и перед ними
# отрицание через ~

scores_typ = scores[~((scores < lower) | (scores > upper))]
```

```python
print("Среднее:", np.mean(scores_typ))
print("Медиана:", np.median(scores_typ))
```

```python
# еще раз выведем среднее и медиану по исходному scores

print(mean_)
print(median_)
```

>**Комментарии.** После того, как мы убрали нетипичные значения из выборки (оставили только типичные и сохранили их в `scores_typ`), среднее значение изменилось более сильно, чем медиана, потому что медиана является устойчивой оценкой центра распределения (устойчивой мерой центральной тенденции), а значит, не зависит от наблюдений на концах выборки, упорядоченной по возрастанию.

*Подсказка:* функции `var()` и `std()` из `np`, для вычисления оценки с $n-1$ в знаменателе нужно добавить аргумент `ddof = 1`.

>Зная среднее и стандартное отклонение, мы можем примерно прикинуть, в каком интервале должны лежать самые частые значения показателя (если нет существенных расхождений между медианой и средним). Самые частые значения лежат в интервале среднее $\pm$ стандартное отклонение. Так, округленное среднее по Орегону 88, округленное стандартное отклонение равно 6. Значит, самые частые значения индекса лежат в интервале 88 $\pm$ 6, то есть от 82 до 94. Для Луизианы получаем 84 $\pm$ 13, то есть самые частые значения индекса лежат в интервале от 71 до 97. Стандартное отклонение позволяет получить более детальную информацию о распределении значений, чем просто среднее или медиана.
